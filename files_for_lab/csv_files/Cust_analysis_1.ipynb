```python
# Customer analysis round 1
# Read the three files into python as dataframes
import pandas as pd  
import numpy as np    # perhaps for later use
# Reading data
data1 = pd.read_csv('file1.csv')
data2 = pd.read_csv('file2.csv')
data3 = pd.read_csv('file3.csv')
data1 = pd.DataFrame(data1)
data2 = pd.DataFrame(data2)
data3 = pd.DataFrame(data3)
```

```python
# Show the DataFrame's shape
print('Data1:', data1.shape)
print('Data2:', data2.shape)
print('Data3:', data3.shape)
```

```python
# Standardize header names
# In work situation I would first rearrange and then combine and take 
# one version of column name as standard. For the sake of the Lab:
# Step one: standardize in upper
# Step two: change name of some columns ST into STATE
# DF 1
cols1 = []
for column in data1.columns:
    cols1.append(column.upper())
data1.columns = cols1
# DF 2
cols2 = []
for column in data2.columns:
    cols2.append(column.upper())
data2.columns = cols2
# DF 3
cols3 = []
for column in data3.columns:
    cols3.append(column.upper())
data3.columns = cols3
# Renaming ST column
data1 = data1.rename(columns={"ST":"STATE"})
data2 = data2.rename(columns={"ST":"STATE"})
# display(data1.head())
# display(data2.head())
# display(data3.head())
```

```python
# Rearrange the columns in the dataframe as needed
# Step 1: determine the desired order
# Step 2: change into that order
# I take data1 as desired order
column_names = data1.columns
data2 = data2.reindex(columns=column_names)
data3 = data3.reindex(columns=column_names)
# display(data1.head())
# display(data2.head())
# display(data3.head())
```

```python
# Concatenate the three dataframes into a dataframe with meaningful name
# Checking rows beforehand for testing
# print('Data1:', data1.shape)
# print('Data2:', data2.shape)
# print('Data3:', data3.shape)
data = [data1, data2, data3]
donors = pd.concat(data)
donors
```
